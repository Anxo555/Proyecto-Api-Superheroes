```python
import requests
import pandas as pd

Clave_Acceso = "82f72b70efe3582c843f7c2ea378986b"

def obtener_heroes_completo(max_id=50, valor_faltante="Sin dato"):
    heroes_data = []

    for hero_id in range(1, max_id + 1):
        url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}"
        response = requests.get(url)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get("response") == "success":
                # Estadísticas
                stats = data.get("powerstats", {})
                # Apariencia
                appearance = data.get("appearance", {})
                # Biografía
                biography = data.get("biography", {})

                def parse_stat(value):
                    """Convierte a entero o retorna valor_faltante si no es un número válido"""
                    try:
                        return int(value)
                    except (ValueError, TypeError):
                        return valor_faltante

                hero = {
                    "ID": data.get("id", valor_faltante),
                    "Nombre": data.get("name", valor_faltante),
                    
                    # Estadísticas convertidas a enteros
                    "Inteligencia": parse_stat(stats.get("intelligence", valor_faltante)),
                    "Fuerza": parse_stat(stats.get("strength", valor_faltante)),
                    "Velocidad": parse_stat(stats.get("speed", valor_faltante)),
                    "Durabilidad": parse_stat(stats.get("durability", valor_faltante)),
                    "Potencia": parse_stat(stats.get("power", valor_faltante)),
                    "Combate": parse_stat(stats.get("combat", valor_faltante)),
                    
                    # Apariencia
                    "Genero": appearance.get("gender", valor_faltante),
                    "Etnia": appearance.get("race", valor_faltante),
                    "Altura_imperial": appearance.get("height", ["Sin dato","Sin dato"])[0],
                    "Altura_metric": appearance.get("height", ["Sin dato","Sin dato"])[1],
                    "Peso_imperial": appearance.get("weight", ["Sin dato","Sin dato"])[0],
                    "Peso_metric": appearance.get("weight", ["Sin dato","Sin dato"])[1],
                    "Color_Ojos": appearance.get("eye-color", valor_faltante),
                    "Color_Pelo": appearance.get("hair-color", valor_faltante),
                    
                    # Biografía
                    "Nombre Completo": biography.get("full-name", valor_faltante),
                    "Alter Egos": biography.get("alter-egos", valor_faltante),
                    "Alias": ", ".join(biography.get("aliases", [])) if biography.get("aliases") else valor_faltante,
                    "Lugar de nacimiento": biography.get("place-of-birth", valor_faltante),
                    "Primera aparición": biography.get("first-appearance", valor_faltante),
                    "Editor": biography.get("publisher", valor_faltante),
                    "Alineación": biography.get("alignment", valor_faltante)
                }

                heroes_data.append(hero)
            else:
                print(f"No se encontró héroe con ID {hero_id}")
        else:
            print(f"Error HTTP {response.status_code} para ID {hero_id}")

    df_heroes = pd.DataFrame(heroes_data)
    df_heroes_completo = obtener_heroes_completo(max_id=50, valor_faltante="Sin dato")
    return df_heroes
   
```
