```python
import numpy as np
import requests
import json
import pandas as pd
from pprint import pprint
from time import sleep
import matplotlib
import matplotlib.pyplot as plt
import random
import seaborn as sns
Clave_Acceso = "82f72b70efe3582c843f7c2ea378986b"
```

```python
def obtener_Estadisticas(max_id=11):
    heroes_data = []

    for hero_id in range(1, max_id + 1):
        url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}"
        response = requests.get(url)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get("response") == "success":
                hero = {
                    "ID": data.get("id", ""),
                    "Nombre": data.get("name", ""),
                    "Inteligencia": data.get("powerstats", {}).get("intelligence", ""),
                    "Fuerza": data.get("powerstats", {}).get("strength", ""),
                    "Velocidad": data.get("powerstats", {}).get("speed", ""),
                    "Durabilidad": data.get("powerstats", {}).get("durability", ""),
                    "Potencia": data.get("powerstats", {}).get("power", ""),
                    "Combate": data.get("powerstats", {}).get("combat", ""),
                    "Nombre Completo": data.get("biography", {}).get("full-name", ""),
                    
                }
                heroes_data.append(hero)
            else:
                print(f"No se encontró héroe con ID {hero_id}")
        else:
            print(f"Error HTTP {response.status_code} para ID {hero_id}")

    # Convertimos la lista de diccionarios en un DataFrame
    df_heroes = pd.DataFrame(heroes_data)
    return df_heroes

# Ejecutamos la función para los primeros 10 héroes
df_heroes = obtener_heroes(max_id=11)

df_heroes

```

```python
def obtener_apariencia_heroes(max_id=10):
    heroes_apariencia = []

    for hero_id in range(1, max_id + 1):
        url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}/appearance"
        response = requests.get(url)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get("response") == "success":
                hero_appearance = {
                    "ID": data.get("id", ""),
                    "Nombre": data.get("name", ""),
                    "Genero": data.get("gender", ""),
                    "Etnia": data.get("race", ""),
                    "Altura_imperial": data.get("height", ["", ""])[0],
                    "Altura_metric": data.get("height", ["", ""])[1],
                    "Peso_imperial": data.get("weight", ["", ""])[0],
                    "Peso_metric": data.get("weight", ["", ""])[1],
                    "Colo_Ojos": data.get("eye-color", ""),
                    "Color_Pelo": data.get("hair-color", "")
                }
                heroes_apariencia.append(hero_appearance)
            else:
                print(f"No se encontró apariencia para ID {hero_id}")
        else:
            print(f"Error HTTP {response.status_code} para ID {hero_id}")

    # Convertimos la lista de diccionarios en un DataFrame
    df_apariencia = pd.DataFrame(heroes_apariencia)
    return df_apariencia

# Ejecutamos la función para los primeros 10 héroes 
df_apariencia = obtener_apariencia_heroes(max_id=10)
df_apariencia



```

```python
def sacar_datos(hero_id):
   
    url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}/biography"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if data.get("response") == "success":
            return data
        else:
            print(f"No se encontró biografía para ID {hero_id}: {data.get('error')}")
            return None
    else:
        print(f"Error HTTP {response.status_code} para ID {hero_id}")
        return None

datos_batman = sacar_datos(500)  
print(datos_batman)
```

```python
def sacar_biografia(hero_id):
   
    url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}/biography"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if data.get("response") == "success":
            return data
        else:
            print(f"No se encontró biografía para ID {hero_id}: {data.get('error')}")
            return None
    else:
        print(f"Error HTTP {response.status_code} para ID {hero_id}")
        return None


ids_heroes = list(range(1, 50))


biografias = []

for hero_id in ids_heroes:
    datos = sacar_biografia(hero_id)
    if datos:
        # Extraemos solo los campos que nos interesan
        biografia = {
            "id": datos.get("id", ""),
            "Nombre": datos.get("name", ""),
            "Nombre Completo": datos.get("full-name", ""),
            "alter-egos": datos.get("alter-egos", ""),
            "Alias": ", ".join(datos.get("aliases", [])),
            "Lugar de nacimiento": datos.get("place-of-birth", ""),
            "Primera aparición": datos.get("first-appearance", ""),
            "Editor": datos.get("publisher", ""),
            "Alineación( El bueno / El malo)": datos.get("alignment", "")
        }
        biografias.append(biografia)


df_biografias = pd.DataFrame(biografias)
df_biografias

```
