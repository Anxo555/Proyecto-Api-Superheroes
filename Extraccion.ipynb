```python
import numpy as np
import requests
import json
import pandas as pd
from pprint import pprint
from time import sleep
import matplotlib
import matplotlib.pyplot as plt
import random
import seaborn as sns
import sys
Clave_Acceso = "82f72b70efe3582c843f7c2ea378986b"
```

```python
def obtener_Estadisticas(max_id=11, valor_faltante="Desconocido"):
    heroes_data = []

    for hero_id in range(1, max_id + 1):
        url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}"
        response = requests.get(url)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get("response") == "success":
                hero = {
                    "ID": data.get("id", ""),
                    "Nombre": data.get("name", ""),
                    "Inteligencia": data.get("powerstats", {}).get("intelligence", ""),
                    "Fuerza": data.get("powerstats", {}).get("strength", ""),
                    "Velocidad": data.get("powerstats", {}).get("speed", ""),
                    "Durabilidad": data.get("powerstats", {}).get("durability", ""),
                    "Potencia": data.get("powerstats", {}).get("power", ""),
                    "Combate": data.get("powerstats", {}).get("combat", ""),
                    "Nombre Completo": data.get("biography", {}).get("full-name", ""),
                    
                }
                heroes_data.append(hero)
            else:
                print(f"No se encontr贸 h茅roe con ID {hero_id}")
        else:
            print(f"Error HTTP {response.status_code} para ID {hero_id}")

    # Convertimos la lista de diccionarios en un DataFrame
    df_heroes = pd.DataFrame(heroes_data)
    df_heroes = df_heroes.replace("null", valor_faltante).fillna(valor_faltante)
    return df_heroes

# Ejecutamos la funci贸n para los primeros 10 h茅roes y templazamos los Nan y null por("Sin Dato").
df_heroes = obtener_Estadisticas(max_id=11, valor_faltante="Sin dato")
df_heroes

```

```python
def obtener_apariencia_heroes(max_id=731):
    heroes_apariencia = []

    for hero_id in range(1, max_id + 1):
        url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}"
        response = requests.get(url)
        
        if response.status_code == 200:
            data = response.json()
            
            if data.get("response") == "success":
                appearance = data.get("appearance", {})

                hero_appearance = {
                    "ID": data.get("id", ""),
                    "Nombre": data.get("name", ""),
                    "Genero": appearance.get("gender", ""),
                    "Etnia": appearance.get("race", ""),          
                    "Altura_metric": appearance.get("height", ["Sin dato", "Sin dato"])[1] if len(appearance.get("height", [])) > 1 else "Sin dato",
                    "Peso_metric": appearance.get("weight", ["Sin dato", "Sin dato"])[1]if len(appearance.get("weight", [])) > 1 else "Sin dato",
                    "Color_Ojos": appearance.get("eye-color", ""),
                    "Color_Pelo": appearance.get("hair-color", "")
                }
                heroes_apariencia.append(hero_appearance)
            else:
                print(f"No se encontr贸 apariencia para ID {hero_id}")
        else:
            print(f"Error HTTP {response.status_code} para ID {hero_id}")

    df_apariencia = pd.DataFrame(heroes_apariencia)

    #reemplazar NaN y "null":
    df_apariencia = df_apariencia.replace("null", "Sin dato").fillna("Sin dato")

    return df_apariencia


# Ejecutamos la funci贸n para los primeros 10 h茅roes 
df_apariencia = obtener_apariencia_heroes(max_id=731)
df_apariencia



```

```python
def sacar_datos(hero_id):
   
    url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}/biography"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if data.get("response") == "success":
            return data
        else:
            print(f"No se encontr贸 biograf铆a para ID {hero_id}: {data.get('error')}")
            return None
    else:
        print(f"Error HTTP {response.status_code} para ID {hero_id}")
        return None

datos_batman = sacar_datos(500)  
print(datos_batman)
```

```python
def sacar_biografia(hero_id):
   
    url = f"https://superheroapi.com/api/{Clave_Acceso}/{hero_id}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        if data.get("response") == "success":
            return data.get("biography", {})
        else:
            print(f"No se encontr贸 biograf铆a para ID {hero_id}: {data.get('error')}")
            return None
    else:
        print(f"Error HTTP {response.status_code} para ID {hero_id}")
        return None


# IDs a consultar
ids_heroes = list(range(1, 100))

biografias = []

for hero_id in ids_heroes:
    datos = sacar_biografia(hero_id)
    if datos:
        biografia = {
            "ID": hero_id,
            "Nombre": datos.get("name", ""),
            "Nombre Completo": datos.get("full-name", ""),
            "Alias": ", ".join(datos.get("aliases", [])),
            "Lugar de nacimiento": datos.get("place-of-birth", ""),
            "Primera aparici贸n": datos.get("first-appearance", ""),
            "Editor": datos.get("publisher", ""),
            "Alineaci贸n (bueno/malo)": datos.get("alignment", "")
        }
        biografias.append(biografia)


df_biografias = pd.DataFrame(biografias)

 #remplazar null/NaN
df_biografias = df_biografias.replace("null", "Sin dato").fillna("Sin dato")

df_biografias

```
